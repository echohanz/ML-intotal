Q2

```python
k = 2
p = 1
score = 0
best = []
keep = True
while p != 6:
    Z, mu, ssd = ml.cluster.kmeans(X, K=k, init='random', max_iter=100)
    if score<=ssd:
        score = ssd
        best.append((Z, mu, ssd))
    p += 1
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
Z, mu, ssd = best[0]
ml.plotClassify2D(None, X, Z)
ax.scatter(mu[:, 0], mu[:, 1], s=400, marker='*', facecolor='black', lw=2)
plt.show()

k = 5
p = 1
score = 0
best = []
keep = True
while p != 6:
    Z, mu, ssd = ml.cluster.kmeans(X, K=k, init='random', max_iter=100)
    if score<=ssd:
        score = ssd
        best.append((Z, mu, ssd))
    p += 1
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
Z, mu, ssd = best[0]
ml.plotClassify2D(None, X, Z)
ax.scatter(mu[:, 0], mu[:, 1], s=400, marker='*', facecolor='black', lw=2)
plt.show()

k = 20
p = 1
score = 0
best = []
keep = True
while p != 6:
    Z, mu, ssd = ml.cluster.kmeans(X, K=k, init='random', max_iter=100)
    if score<=ssd:
        score = ssd
        best.append((Z, mu, ssd))
    p += 1
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
Z, mu, ssd = best[0]
ml.plotClassify2D(None, X, Z)
ax.scatter(mu[:, 0], mu[:, 1], s=400, marker='*', facecolor='black', lw=2)
plt.show()
```

```python
X = np.genfromtxt("data/faces.txt", delimiter=None)
plt.figure()
img = np.reshape(X[3,:],(24,24))
plt.imshow(img.T , cmap="gray")
```

Q1

```python
mean = np.mean(X,axis = 0,keepdims=True)
X0 = X-mean
plt.figure()
img = np.reshape(X0[3,:],(24,24))
plt.imshow(img.T , cmap="gray")
```

Q2

```python
import scipy.linalg
U, s, V = scipy.linalg.svd(X0, full_matrices=False)
W = U.dot(np.diag(s))
print(W.shape, V.shape)
```

Q3

```python
mse = []
for k in range(1, 31):
    X0h = W[:, :k].dot(V[:k,:])
    mse.append(np.mean((X0 - X0h)**2))
plt.plot(range(1,31),mse)
plt.show()
```

Q4

```python
for i in range(3):
    alpha = 2*np.median(np.abs(W[:,i]))
    aaa = mean + alpha*V[i,:]
    aaa = np.reshape(aaa,(24,24))
    plt.figure()
    plt.imshow(aaa.T, cmap='gray')
    plt.show()
    bbb = mean - alpha*V[i,:]
    bbb = np.reshape(bbb,(24,24))
    plt.imshow(bbb.T, cmap='gray')
    plt.show()
```
