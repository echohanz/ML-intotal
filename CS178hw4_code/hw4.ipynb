# Problem 3

```python
import numpy as np
import matplotlib.pyplot as plt
import mltools as ml

np.random.seed(0)
%matplotlib inline

X = np.genfromtxt('C:/Users/19769/Desktop/uci-cs178-win21/X_train.txt', delimiter=None) 
Y = np.genfromtxt('C:/Users/19769/Desktop/uci-cs178-win21/Y_train.txt', delimiter=None) 
X,Y = ml.shuffleData(X,Y)
```

Q1

```python
print('minimum')
for i in range (0, 14):
    print(np.min(X[:,i]))
print("-------------------------------")
print('maximum')
for i in range (0, 14):
    print(np.max(X[:,i]))
print("-------------------------------")
print('mean')
for i in range (0, 14):
    print(np.mean(X[:,i]))
print("-------------------------------")
print('variance')
for i in range (0, 14):
    print(np.var(X[:,i]))
```

Q2

```python
Xtr = X[0:10000]
Ytr = Y[0:10000]
Xva = X[10000:20000]
Yva = Y[10000:20000]
learner = ml.dtree.treeClassify(Xtr,Ytr,maxDepth=50)
ter = learner.err(Xtr, Ytr)
ver = learner.err(Xva, Yva)
print(ter, ver)
```

Q3

```python
ter1 = []
ver1 = []
for i in range(0, 16):
    learner1 = ml.dtree.treeClassify(Xtr,Ytr,maxDepth=i)
    ter1.append(learner1.err(Xtr, Ytr))
    ver1.append(learner1.err(Xva, Yva))
    
f, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.plot(ter1,c='r')
ax.plot(ver1,c='b')
plt.show()
```

Models with higher maxDepth have higher complexity and maxDepth 4 provies the best decision tree model.

Q4

```python
ter2 = []
ver2 = []
minParent = []
for i in range(2, 13):
    learner2 = ml.dtree.treeClassify(Xtr,Ytr,maxDepth=50,minParent=2**i)
    ter2.append(learner2.err(Xtr, Ytr))
    ver2.append(learner2.err(Xva, Yva))
    minParent.append(2**i)


f, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.plot(minParent,ter2,c='r')
ax.plot(minParent,ver2,c='b')
plt.show()
```

Models with higher minParent have lower complexity and minParent of 2048 provides the best decision tree model.

Q5

```python
learner3 = ml.dtree.treeClassify(Xtr,Ytr,maxDepth=4)
troc = learner3.roc(Xtr, Ytr)
vroc = learner3.roc(Xva, Yva)
f, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.plot(troc, c='r')
ax.plot(vroc, c='b')
print(learner3.auc(Xtr, Ytr), learner3.auc(Xva, Yva))
```
